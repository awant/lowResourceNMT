```python
with open('/data/LowResourceTmp/en_dict', 'rb') as f:
    en_dict = f.read()
```

```python
en_dict = pickle.loads(en_dict)[0]
```

```python
top_words = en_dict.most_common(5000)
```

```python
from translate import Translator
from tqdm import tqdm

engine = Translator(to_lang='ru')

en_ru_dict = dict()


for (word, _) in tqdm(top_words, position=0):
    try:
        translation = engine.translate(word)
        en_ru_dict[word] = translation
    except Exception:
        pass
```

```python
en_ru_dict
```

```python
from nltk.stem.snowball import RussianStemmer, EnglishStemmer
```

```python
stemmer_en = EnglishStemmer()
stemmer_ru = RussianStemmer()
```

```python
en_ru_stemmed = {stemmer_en.stem(key): stemmer_ru.stem(val.split(' ')[0]) for key, val in en_ru_dict.items()}
```

```python
en_ru_stemmed
```

```python
with open('en-ru-dict', 'wb') as f:
    f.write(pickle.dumps(en_ru_stemmed))
```
